AND 연산을 수행하는 뉴런 생성  
___
AND 연산은 여러개의 입력을 받을 수 있지만 지금은 2개의 입력으로 제한

AND 연산은 입력이 모두 참 값일 때 참이 되고, 그 밖의 경우는 모두 거짓이 된다  
딥러닝의 주요 입력값은 정수나 실수이다  
아래의 코드로 확인해보자 
         

```
print(int(True))
print(int(False))
```

True와 False가 각각 1과 0으로 나타나는것을 알 수 있다

하나의 뉴런은 여러 개의 입력을 받을 수 있기 때문에 입력이 2개여도 문제가 없다  
다만 편향은 보통 하나만 사용한다

입력치와 가중치를 곱한 다음 서로 더하고, 활성화 함수에 넣으면 뉴런의 출력을 계산할 수 있다

```
import math
import tensorflow as tf
import numpy as np

def sigmoid(x):
  return 1/ (1 + math.exp(-x))

x = np.array([[1,1], [1,0],[0,1],[0,0]])
y = np.array([[1],[0],[0],[0]])
w = tf.random.normal([2],0,1)
b = tf.random.normal([1],0,1)
b_x = 1

for i in range(2000):
  error_sum = 0
  for j in range(4):
    output = sigmoid(np.sum(x[j]*w)+b_x*b)
    error = y[j][0] - output
    w = w + x[j] * 0.1 * error
    b = b + b_x * 0.1 * error
    error_sum += error
  
  if i % 200 == 199:
    print(i, error_sum)
```
