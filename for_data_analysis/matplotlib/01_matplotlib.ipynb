## matplotlib로 그래프 그리기

````python
import matplotlib.pyplot as plt
````

### 선 그래프

`plt.plot([x,]y[,fmt]`

fmt는 format string으로 다양한 형식으로 그래프를 그릴 수 있는 옵션

x와 y는 x축과 y축을 나타낸다  

````python
data1 = [10, 14, 19, 20, 25]
````

````python
plt.plot(data1)
````

객체 정보 없이 그래프만 출력하는 코드

````python
plt.plot(data1)
plt.show()
````

x값과 y값이 모두 있는 데이터로 그래프 그리기

````python
import numpy as np
````

````python
x = np.arange(-4.5,5,0.5) # 배열 x 생성,  범위 : (-4.5 ~ 5), 0.5씩 증가
y = 2*x**2 # 수식을 이용해 배열 x에 대응하는 배열 y 생성
[x,y]
````

````python
plt.plot(x,y)
plt.show()
````

### 여러 그래프

#### 여러 개의 데이터를 한번에 그리기

연속해서 plot()을 수행해도 되고
```
plt.plot([x1,]y1[,fmt1])
plt.plot([x1,]y2[,fmt2])
....
```
하나의 plot()에 데이터를 모두 입력하는 방법으로도 가능하다  
```
plt.plot(x1,y1[,fmt1],x2,y2[,fmt2],...)
```


````python
x = np.arange(-4.5,5,0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10
````

````python
# plot() 을 여러번 실행하는 방식
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
````

````python
# 하나의 plot()에 데이터를 모두 사용한 방식
plt.plot(x, y1, x, y2, x, y3)
````

하나의 그래프 창에 여러개를 그리는 대신  
새로운 그래프 창에 각각 그리는 방법  
`plt.figure()`

````python
plt.plot(x ,y1)

plt.figure()

plt.plot(x, y2)

plt.show()
````

그래프 창의 번호를 지정해 그래프를 그릴 수도 있다

`plt.figure(n)`

````python
x = np.arange(-5,5,0.1)
y1 = x**2 - 2
y2 = 20*np.cos(x)**2

plt.figure(1)
plt.plot(x, y1)

plt.figure(2)
plt.plot(x, y2)

plt.figure(1)
plt.plot(x, y2)

plt.figure(2)
plt.clf() # 2번 그래프창에 그려진 모든 그래프 삭제 
plt.plot(x, y1)

plt.show()
````

#### 그래프 창 나누기

하나의 그래프 창을 하위 그래프 영역으로 나누고,  
각 영역에 그래프 그리기  
```
plt.subplot(m, n, p)
```  
m x n 행렬로 이루어진 하위 그래프 중,  
p번 위치에 그래프가 그려지도록 지정

````python
x = np.arange(0,10,0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1

plt.subplot(2,2,1)
plt.plot(x, y1)

plt.subplot(2,2,2)
plt.plot(x, y2)

plt.subplot(2,2,3)
plt.plot(x, y3)

plt.subplot(2,2,4)
plt.plot(x, y4)

plt.show()
````

#### 그래프 출력 범위

x축과 y축의 범위를 지정해 그릴 수 있다
```
plt.xlim(xmin, xmax) # x축 범위 지정
plt.ylim(ymin, ymax) # y축 범위 지정
```

````python
x = np.linspace(-4,4,100) # -4 ~ 4 범위에서 100개 생성
y1 = x**3
y2 = 10*x**2 - 2

plt.plot(x,y1, x,y2)
plt.show()
````

y1과 y2 그래프가 만나는 지점이 잘 보이지 않으므로  
출력범위를 지정해 확대 

````python
plt.plot(x,y1, x,y2)
plt.xlim(-1,1)
plt.ylim(-3,3)
plt.show()
````
