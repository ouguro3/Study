### 출력 형식 지정

`fmt` 옵션을 이용해 그래프에서 선의 출력 형식을 지정하고,  
축의 라벨, 제목, 격자, 문자열을 추가할 수 있다
```
fmt = '[color][line_style][maker]'
```

**컬러 지정**

|컬러 약어|컬러|            
|---------|----|
|b|파란색|
|g|녹색|
|r|빨간색|
|c|청녹색|
|m|자홍색|
|y|노란색|
|k|검은색|
|w|흰색|



**선 스타일 지정**

|선 스타일 약어|선 스타일|            
|---------|----|
|-|실선|
|--|파선|
|:|점선|
|-.|파선,점선 혼합선|

마커 지정

|마커 약어|마커|            
|---------|----|
|o|원 모양|
|^,v,<,>|삼각형 위쪽, 아래쪽, 왼쪽, 오른쪽|
|s|사각형|
|p|오각형|
|h,H|육각형1,육각형2|
|*|별 모양|
|+|더하기|
|x,X|x, 채워진x|
|D,d|다이아몬드, 얇은 다이아몬드|

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.arange(0,5,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
````

**옵션 지정하지 않은 선 그래프**

````python
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
````

**색 지정**

````python
plt.plot(x, y1, 'm', x, y2, 'y', x, y3, 'k', x, y4, 'c')
plt.show()
````

**선 스타일 지정**

````python
plt.plot(x, y1, '-', x, y2, '--', x, y3, ':', x, y4, '-.')
plt.show()
````

**마커 지정**

````python
plt.plot(x, y1, 'o', x, y2, '^', x, y3, 'v', x, y4, 'd')
plt.show()
````

**컬러, 선 스타일, 마커 혼합 지정**

````python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.show()
````

### 라벨, 제목, 격자, 범례, 문자열 표시

**x 축 라벨 추가 :** `xlabel('문자열')`  
**y 축 라벨 추가 :** `ylabel('문자열')`

````python
x = np.arange(-4.5, 5, 0.5)
y = 2*x**3

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

**그래프 제목 추가 :** `title('문자열')`

````python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.show()
````

**격자 추가 :** `grid(True) or grid()`

````python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.grid(True) # or grid()
````

**범례 추가 :** `legend([str1. str2...])` 

````python
x = np.arange(0,5,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x , y1, '>-r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['data1','data2','data3','data4'])
plt.show()
````

`legend()` 의 `loc = 위치 코드 or 위치 문자열` **옵션으로 범례의 위치를 지정할 수 있다**


|범례 위치|위치 문자열|위치 코드|
|---------|:-----------:|:---------:|
|최적 위치 자동 선정|best|0|
|상단 우측|upper right|1|
|상단 좌측|upper left|2|
|하단 좌측|lower left|3|
|하단 우측|lower right|4|
|중앙|right|5|
|중앙 좌측|center left|6|
|중앙 우측|center right|7|
|하단 중앙|lower center|8|
|상단 중앙|upper center|9|
|중앙|best|10|



````python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['data1', 'data2','data3','data4'], loc='lower right')
plt.show()
````

````python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['data1', 'data2','data3','data4'], loc=4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.grid()
plt.show()
````
